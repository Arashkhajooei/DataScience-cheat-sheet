```python
import random

locations = ['3', '-6', '1']
weights = [0.5, 0.2, 0.3]

locations2 = ['0', '9', '-12', '1', '-2']
weights2 = [0.2, 0.2, 0.1, 0.3, 0.2]


class Person:
    def __init__(self):
        self.position = 0

    def update_position(self, value):
        # Update the person's position based on the given value,
        # ensuring that the position does not go below 0.
        self.position = max(self.position + int(value), 0)


def print_person_location(person_name, position):
    # Print the person's location as dashes ('-') representing the units of location.
    print(f"{person_name}'s location:", "-" * position, f"({position})")


def simulate_game():
    # Create instances of Person for both players.
    person1 = Person()
    person2 = Person()

    iterations = 0
    while person1.position < 120 and person2.position < 120:
        # Randomly select a location update for each person based on weights.
        Person1_location = random.choices(locations, weights)[0]
        Person2_location = random.choices(locations2, weights2)[0]

        # Update the positions of both players based on the selected locations.
        person1.update_position(Person1_location)
        person2.update_position(Person2_location)

        iterations += 1
        print(f"----- Iteration {iterations} -----")
        print_person_location("Person1", person1.position)
        print_person_location("Person2", person2.position)

    print("\nNumber of iterations:", iterations)
    if person1.position >= 120:
        print("Person1 is the winner!")
    else:
        print("Person2 is the winner!")

    # Uncomment the following lines if you want to print the final positions of both players.
    # print_person_location("Person1", person1.position)
    # print_person_location("Person2", person2.position)


simulate_game()
```
