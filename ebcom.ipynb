```python
def swap_rows(df, i1, i2):
    a, b = df.iloc[i1, :].copy(), df.iloc[i2, :].copy()
    df.iloc[i1, :], df.iloc[i2, :] = b, a
    return df

swap_rows(df_provider, df_provider.loc[df_provider['Metal'].isin(['Gold','Silver'])].index, 168)
```
